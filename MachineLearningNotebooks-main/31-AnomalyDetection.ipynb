```python
fraud_ratio = 0.001727
```

```python
# let's compare models to see if anomaly detection works here
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(
    n_estimators=200, #how many isolation tree?
    contamination=fraud_ratio,
    max_samples="auto", # default auto, how many samples will be in each tree?
    max_features=1.0, #all features
    random_state=42,
    n_jobs=-1
)
```

```python
X = df.drop("Class", axis=1)

iso.fit(X)
```

```python
df["iso_raw"] = iso.predict(X)
```

```python
df['iso_raw'].unique()
```

```python
#1  → normal
#-1 → anomaly
```

```python
df["iso_pred"] = df["iso_raw"].map({1:0, -1:1})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(["Class","iso_raw","iso_pred"], axis=1)
y = df["Class"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,
    random_state=42
)
```

```python
iso_test_pred = df.loc[X_test.index, "iso_pred"]
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, iso_test_pred))
```

```python
# recall seems a little bit low with 24% but considering only 0.17% of the whole data could be anomaly
# it can be actually good. now let's look at what we should actually look at
```

```python
df["anomaly_score"] = iso.decision_function(X)
```

```python
sns.kdeplot(
    df[df["Class"]==0]["anomaly_score"],
    label="Normal"
)

sns.kdeplot(
    df[df["Class"]==1]["anomaly_score"],
    label="Fraud"
)

plt.legend()
plt.title("Anomaly Score Distribution")
plt.show()
```

```python
# anomaly score is distributed to left in anomalies and to right in normal which should be exactly like this
```

```python
# let's see top 1% suspicious transaction
```

```python
df["anomaly_score"].quantile(0.01)
```

```python
# if we increase threshold to this rather than 0.0017 let's see what happens
```

```python
threshold = df["anomaly_score"].quantile(0.01)

df["iso_custom"] = (df["anomaly_score"] < threshold).astype(int)
```

```python
print(classification_report(df["Class"], df["iso_custom"]))
```

```python
# as you can see recall actually increases and precision drops 1%. this is what we want in real life
```

```python
sns.kdeplot(df[df["Class"]==0]["anomaly_score"], label="Normal")
sns.kdeplot(df[df["Class"]==1]["anomaly_score"], label="Fraud")

plt.axvline(threshold, color="red", linestyle="--",
            label="Decision Threshold")

plt.legend()
plt.title("Anomaly Score Separation")
plt.show()
```

```python
#in isolation forest, model learns about decision boundry. it produces suspicion score
#however business makes the decision here like this to determine whether something is fradulent or not
```

```python
# in real life banks get millions of transaction everyday
# so they just really want to send top 1% suspicious transactions for evaluation anyway
```

```python
# if we sort values accordingly
```

```python
df_sorted = df.sort_values("anomaly_score")
```

```python
# then we get top 1%
k = int(len(df_sorted) * 0.01)

top_k = df_sorted.head(k)
```

```python
frauds_captured = top_k["Class"].sum()
total_frauds = df["Class"].sum()

recall_at_k = frauds_captured / total_frauds

print(recall_at_k)
```

```python
# it means that just by evaluating 1% of transactions we recovered almost 60% of fraud
```

```python
# we can expand it like this
```

```python
percentages = np.linspace(0.001, 0.05, 30)
recalls = []

total_frauds = df["Class"].sum()

for p in percentages:
    k = int(len(df_sorted) * p)
    captured = df_sorted.head(k)["Class"].sum()
    recalls.append(captured / total_frauds)
```

```python
plt.plot(percentages*100, recalls)

plt.xlabel("Investigated Transactions (%)")
plt.ylabel("Fraud Recall")
plt.title("Top-K Capture Curve (Isolation Forest)")

plt.show()
```

```python
# so by investigating 5% of total transactions you can get more than 80% of fradulent ones
```

```python
# is this enough for a real life case? probably not.
# in real life banks try to feed this info learnt from unsupervised learning to supervised learning
# we can use anomaly_score in the df to give a solid feature for supervised learning algorithm. 
# let's see how it works
```

```python
from xgboost import XGBClassifier
```

```python
X = df.drop(["Class", "iso_raw", "iso_pred", "iso_custom"], axis=1)
# anomaly_score is already in X 

y = df["Class"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

xgb = XGBClassifier(
    scale_pos_weight=len(y_train[y_train==0]) / len(y_train[y_train==1]),
    random_state=42, n_jobs=-1
)
#scale_pos_weight to handle unbalanced dataset

xgb.fit(X_train, y_train)
print(classification_report(y_test, xgb.predict(X_test)))
```
